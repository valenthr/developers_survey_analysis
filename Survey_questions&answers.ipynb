```python
public = pd.read_csv('C:/Users/User/Downloads/survey_results_public.csv')
schema = pd.read_csv('C:/Users/User/Downloads/survey_results_schema.csv')
```

```python
count = public['ResponseId'].count() #How many participants in the survey?
print(count)
```

```python
questions = set(schema['qname'].unique())&set(public.columns) # How many respondents answered all questions?
public_questions = public.dropna(subset=questions)
compl_count =  public_questions['ResponseId'].count()
print(f'{compl_count} users answered all questions')
```

```python
e_mean = public['WorkExp'].mean() #what is the average experience?
e_median = public['WorkExp'].median() #what is the median experience?
e_mode = public['WorkExp'].mode() #what is the mode in experience?
print(f'Mean = {e_mean}, Median = {e_median}, Mode = {e_mode}')
```

```python
remote_workers = len(public[public['RemoteWork']=='Remote'])
perc_remote = remote_workers/ public['RemoteWork'].count() * 100 # How many respondents work remotely?
print(f'{perc_remote}% workers working remotely')
```

```python
python_users = public['LanguageHaveWorkedWith'].str.contains('Python', na=False) #How many % use Python?
py_perc = python_users.sum() / public['ResponseId'].count() * 100
print(f"{py_perc}% respondents use Python")
```

```python
compl_cources = public['LearnCodeOnline'].str.contains('Online Courses',case=False , na=False).sum() #How many respondents learned using online courses?
cour_perc = compl_cources/public['LearnCodeOnline'].count()*100
print(f'{cour_perc}% responders used online courses to learn')
```

```python
# Which is top 10 countries by salary of developers, who program in Python?
python_users = public['LanguageHaveWorkedWith'].str.contains('Python', na=False)
python_comp = (public[python_users].dropna(subset=['ConvertedCompYearly']).groupby('Country', as_index=False).agg({'ConvertedCompYearly':['mean', 'median']}).droplevel(1, axis = 1))
python_comp.columns = (['country', 'mean', 'median'])
python_comp = python_comp.astype({'mean':'int','median':'int'})
python_comp = python_comp.sort_values(by='median', ascending = False)
python_comp.to_csv()
print(python_comp.head(10))

#What levels of education do the 5 respondents with the highest compensation have?
top_5_education = public.nlargest(5, 'ConvertedCompYearly')['EdLevel']
print(f"Degree of 5 the largest paid respondents: {top_5_education}")
```

```python
# What percentage of respondents program in Python in each age category?
all_us = public.groupby('Age')['ResponseId'].count()
pyth_us = public[python_users].groupby('Age')['ResponseId'].count()
per_python_dev = (pyth_us/all_us * 100).fillna(0)
print(f"Percentage of python developers by age {per_python_dev}")
```

```python
# Among respondents who are in the 75th percentile for compensation and work remotely, which industries are most common?
high_income_remote = public[(public['ConvertedCompYearly'] >= public['ConvertedCompYearly'].quantile(0.75)) & (public['RemoteWork'] == 'Remote')]
top_industries = high_income_remote['Industry'].value_counts().head()
print(f"Number of developers with salaries in third quartile by industry {top_industries}") 
```
